```python
x=[1,2,3,4,5]
y_s=[i*i for i in x]
y_c=[i*i*i for i in x]
```

```python
y_s
```

```python
y_c
```

```python
plt.plot(x,y_s)
plt.plot(x,y_c)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
plt.plot(x,y_s,label='squares')
plt.plot(x,y_c,label='cubes')
plt.plot(x,[2,4,6,8,10])
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc=10)
plt.show()
```
