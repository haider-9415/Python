```python
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import rcParams as rc
import numpy as np
import random as ra
```

```python
marks=[]
for i in range(100):
    x=ra.randint(0,20)
    marks.append(x)
# marks
```

```python
len(marks)
```

```python
plt.hist(marks)
plt.show()
```

```python
rc['figure.figsize']=10,5
```

```python
plt.hist(marks,edgecolor='red',bins=4)
plt.ylabel('students')
plt.xlabel('marks')
plt.title('AI Test Analysis')
y=[0,5,10,15,20,25,30]
plt.yticks(y)
plt.show()
```

```python
b=[0,7,13,17,20]
plt.hist(marks,edgecolor='red',bins=b)
plt.ylabel('students')
plt.xlabel('marks')
plt.title('AI Test Analysis')
y=[0,5,10,15,20,25,30,35,40]
plt.yticks(y)
plt.xticks(b)
plt.show()
```
