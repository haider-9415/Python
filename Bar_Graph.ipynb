```python
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import rcParams as rc
import numpy as np
```

```python
subjects=['Math','Science','English','S.S.','Hindi','A.I.']
marks=[50,71,73,70,52,100]
plt.xlabel('Marks')
plt.ylabel('Subjects')
plt.title('Marksheet')
rc['figure.figsize']=15,10
plt.bar(subjects,marks,width=0.2,bottom=20,color='red')
plt.show();
```

```python
subjects=['Math','Science','English','S.S.','Hindi','A.I.']
marks=[50,71,73,70,52,100]
plt.xticks(arr1,subjects)
plt.xlabel('Marks')
plt.ylabel('Subjects')
plt.title('Marksheet')
rc['figure.figsize']=10,5
plt.bar(arr1,marks)
plt.show();
```

```python
plt.xticks(arr1,subjects)
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.title('Marksheet')
plt.bar(arr1,marks,label=2021)
plt.bar(arr1,marks1,label=2022)
plt.legend()
plt.show();
```

```python
plt.xticks(arr1,subjects)
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.title('Marksheet')
rc['figure.figsize']=10,5
plt.bar(arr1+0.2,marks,width=0.4,label=2021)
plt.bar(arr1-0.2,marks1,width=0.4,label=2022)
plt.legend()
plt.show();
```

```python
plt.barh(arr1+0.2,marks,height=0.4,label=2021,color='red')
plt.barh(arr1-0.2,marks1,height=0.4,label=2022,color='blue')
rc['figure.figsize']=10,5
plt.yticks(arr1,subjects)
plt.ylabel('Subjects')
plt.xlabel('Marks')
plt.title('Marksheet')
plt.legend()
plt.show();
```

```python
plt.xticks(arr1,subjects)
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.title('Marksheet')
plt.bar(arr1,marks,label=2021)
plt.bar(arr1,marks1,bottom=marks,label=2022)
plt.legend()
plt.show();
```
